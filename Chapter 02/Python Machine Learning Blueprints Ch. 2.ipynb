```python
# split using the bullet
def parse_info(row):
        if not 'sqft' in row:
            br, ba = row.split('•')[:2]
            sqft = np.nan
        else:
            br, ba, sqft = row.split('•')[:3]                
        return pd.Series({'Beds': br, 'Baths': ba, 'Sqft': sqft})
```

```python
attr = sucln['propertyinfo_value'].apply(parse_info) # 按照•分割成多个字段
```

```python
attr.head()
```

```python
#remove the strings from our values 删除掉ba、bd sqft等字串
attr_cln = attr.applymap(lambda x: x.strip().split(' ')[0] if isinstance(x,str) else np.nan)

attr_cln.head()
```

```python
sujnd = sucln.join(attr_cln)
```

```python
sujnd.T # 完成了卧室、浴室和平方英尺字段的拆分
```

```python
# parse out zip, floor

def parse_addy(r):
    so_zip = re.search(', NY(\d+)', r) # NY打头后面的数字是邮编
    so_flr = re.search('(?:APT|#)\s+(\d+)[A-Z]+,', r) # APT后面跟随的数字是楼层
    if so_zip:
        zipc = so_zip.group(1)
    else:
        zipc = np.nan
    if so_flr:
        flr = so_flr.group(1)
    else:
        flr = np.nan
    return pd.Series({'Zip':zipc, 'Floor': flr})
```

```python
flrzip = sujnd['routable_link/_text'].apply(parse_addy)
```

## Regression Modeling

```python
su_lt_two.sort_values('Zip')
```

```python
import patsy
import statsmodels.api as sm

# 该行描述输入输出变量之间的关系：我们想知道邮政编码和w卧室数量将如何影响出租价格
f = 'Rent ~ Zip + Beds' 
# 传入公式和数据，输出X矩阵由预测变量组成，y向量由响应变量组成
y, X = patsy.dmatrices(f, su_lt_two, return_type='dataframe')
```

```python
results = sm.OLS(y, X).fit() # 运行模型
print(results.summary())     # 打印模型结果
```
